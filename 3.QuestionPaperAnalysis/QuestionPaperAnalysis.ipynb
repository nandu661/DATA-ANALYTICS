```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("class_marks.csv")
```

```python
df['Q1']=df["Q1aM4"]+df["Q1bM6"]
df['Q2']=df['Q2aM6']+df['Q2bM4']
df['Q3']=df['Q3aM5']+df['Q3bM5']
df['Q4']=df['Q4aM3']+df['Q4bM7']
df['Q5']=df['Q5M10']
df['Q6']=df['Q6aM4']+df['Q6bM6']
df
```

```python
df.fillna(0,inplace=True)
df
```

```python
df=df.drop(["Q1aM4", "Q1bM6","Q2aM6", "Q2bM4","Q3aM5","Q3bM5","Q4aM3","Q4bM7","Q5M10","Q6aM4","Q6bM6"], axis=1)
df
```

```python
data = {
    'Q1': [5, 12, 15, 20, 25],
    'Total': [30, 40, 50, 60, 70]
}
df = pd.DataFrame(data)
df.plot.line(x='Q1', y='Total',linestyle='--', marker='o', color='red')
plt.title("Line Graph of Q1 vs Total")
plt.xlabel("Q1")
plt.ylabel("Total")
plt.show()
```

each marked with a circle. The graph has titles and labels to explain what the axes represent.
